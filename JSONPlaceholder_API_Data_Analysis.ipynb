```python
import requests
import pandas as pd
import json
from IPython.display import display

# Base URL for JSONPlaceholder API
BASE_URL = "https://jsonplaceholder.typicode.com"

def fetch_data(endpoint):
    """Fetch data from specified endpoint"""
    response = requests.get(f"{BASE_URL}/{endpoint}")
    return response.json()

# Fetch sample data from different endpoints
posts = fetch_data("posts")
users = fetch_data("users")
comments = fetch_data("comments")

# Convert to DataFrames for easier analysis
posts_df = pd.DataFrame(posts)
users_df = pd.DataFrame(users)
comments_df = pd.DataFrame(comments)

# Display sample data
print("\nSample Posts Data:")
display(posts_df.head(2))

print("\nSample Users Data:")
display(users_df.head(2))

print("\nSample Comments Data:")
display(comments_df.head(2))

# Print data structure
print("\nData Structure Analysis:")
print("\nPosts columns:", posts_df.columns.tolist())
print("Users columns:", users_df.columns.tolist())
print("Comments columns:", comments_df.columns.tolist())
```
